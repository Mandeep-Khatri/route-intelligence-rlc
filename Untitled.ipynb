## Geographic Distance Analysis

This section geocodes donor addresses to compute real distances between donors, enabling us to validate whether the batching hotspots identified earlier are genuinely close enough to make combined pickups practical.

```python
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import folium
from math import radians, sin, cos, sqrt, atan2
import itertools
import os
```

```python
# --- STEP 1: Extract unique donor addresses ---
# Work with unique donor locations only to avoid redundant geocoding calls
unique_donors = df[['Donor Name', 'Donor Address', 'Donor City', 'Donor State']].drop_duplicates(subset=['Donor Name'])
unique_donors = unique_donors.copy()
unique_donors['Full_Address'] = (
    unique_donors['Donor Address'].str.strip() + ', ' +
    unique_donors['Donor City'].str.strip() + ', ' +
    unique_donors['Donor State'].str.strip() + ', USA'
)
print(f'Unique donors to geocode: {len(unique_donors)}')
unique_donors[['Donor Name', 'Full_Address']].head(10)
```

```python
# --- STEP 2: Geocode donor addresses ---
# Uses OpenStreetMap Nominatim (free, no API key needed)
# Rate-limited to 1 request/second to respect usage policy
# Results are saved to CSV so geocoding only runs once

GEOCODE_CACHE = 'donor_geocoded.csv'

if os.path.exists(GEOCODE_CACHE):
    print('Loading geocoded data from cache...')
    geocoded = pd.read_csv(GEOCODE_CACHE)
else:
    print('Geocoding addresses (this will take ~10 minutes for ~500 addresses)...')
    geolocator = Nominatim(user_agent='food_rescue_team8_v1')
    geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1.1, error_wait_seconds=5)

    def safe_geocode(address):
        try:
            loc = geocode(address)
            if loc:
                return pd.Series([loc.latitude, loc.longitude])
        except Exception as e:
            print(f'  Error geocoding "{address}": {e}')
        return pd.Series([None, None])

    unique_donors[['latitude', 'longitude']] = unique_donors['Full_Address'].apply(safe_geocode)
    geocoded = unique_donors.dropna(subset=['latitude', 'longitude'])
    geocoded.to_csv(GEOCODE_CACHE, index=False)
    print(f'Geocoded and saved {len(geocoded)} locations.')

print(f'Geocoded donors available: {len(geocoded)}')
geocoded[['Donor Name', 'Donor City', 'latitude', 'longitude']].head(10)
```

```python
# --- STEP 3: Haversine distance function ---
# Computes straight-line distance in miles between two lat/lon points

def haversine_miles(lat1, lon1, lat2, lon2):
    R = 3958.8  # Earth radius in miles
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    return R * 2 * atan2(sqrt(a), sqrt(1 - a))

# Quick test
d = haversine_miles(42.3601, -71.0589, 42.3611, -71.0560)  # ~0.15 miles
print(f'Test distance (should be ~0.15 miles): {d:.3f} miles')
```

```python
# --- STEP 4: Compute pairwise distances between donors per city ---
# For each city, compute distance between every pair of donors
# Flag pairs within BATCH_THRESHOLD miles as 'batchable'

BATCH_THRESHOLD = 0.5  # miles — donors within 0.5 miles are candidates for batching

batch_pairs = []

for city, group in geocoded.groupby('Donor City'):
    donors = group.reset_index(drop=True)
    for i, j in itertools.combinations(range(len(donors)), 2):
        d1 = donors.iloc[i]
        d2 = donors.iloc[j]
        dist = haversine_miles(d1['latitude'], d1['longitude'], d2['latitude'], d2['longitude'])
        if dist <= BATCH_THRESHOLD:
            batch_pairs.append({
                'City': city,
                'Donor_A': d1['Donor Name'],
                'Donor_B': d2['Donor Name'],
                'Distance_Miles': round(dist, 3)
            })

batch_df = pd.DataFrame(batch_pairs).sort_values(by=['City', 'Distance_Miles'])
print(f'Total batchable donor pairs (within {BATCH_THRESHOLD} miles): {len(batch_df)}')
batch_df.head(20)
```

```python
# --- STEP 5: Estimate cost savings from batching ---
# For each batchable pair, look up how many times BOTH donors had pickups on the SAME date
# Each same-day co-occurrence is a trip that COULD have been batched into one
# Saving: $15-$20 driver cost per merged trip (use $15 as conservative estimate)

DRIVER_COST_SAVING = 15  # $ saved per merged trip (conservative)

# Merge geocoded lat/lon back to main df
df_geo = df.merge(
    geocoded[['Donor Name', 'latitude', 'longitude']],
    on='Donor Name',
    how='left'
)

# Count same-day co-occurrences for each batchable pair
savings_rows = []

for _, pair in batch_df.iterrows():
    donor_a_dates = set(df_geo[
        (df_geo['Donor Name'] == pair['Donor_A']) &
        (df_geo['Rescue is Finished'] == True)
    ]['Rescue Date'].dt.date)
    
    donor_b_dates = set(df_geo[
        (df_geo['Donor Name'] == pair['Donor_B']) &
        (df_geo['Rescue is Finished'] == True)
    ]['Rescue Date'].dt.date)
    
    overlap_days = len(donor_a_dates & donor_b_dates)
    potential_saving = overlap_days * DRIVER_COST_SAVING
    
    savings_rows.append({
        'City': pair['City'],
        'Donor_A': pair['Donor_A'],
        'Donor_B': pair['Donor_B'],
        'Distance_Miles': pair['Distance_Miles'],
        'Same_Day_Pickups': overlap_days,
        'Potential_Saving_USD': potential_saving
    })

savings_df = pd.DataFrame(savings_rows).sort_values(by='Potential_Saving_USD', ascending=False)

total_saving = savings_df['Potential_Saving_USD'].sum()
print(f'Total estimated savings from batching nearby donors: ${total_saving:,.0f}')
print(f'(Based on {len(savings_df)} batchable pairs, ${DRIVER_COST_SAVING}/trip saved, historical data only)')
savings_df.head(20)
```

```python
# --- STEP 6: Summary by city ---
city_summary = savings_df.groupby('City').agg(
    Batchable_Pairs=('Donor_A', 'count'),
    Total_Same_Day_Pickups=('Same_Day_Pickups', 'sum'),
    Total_Potential_Saving_USD=('Potential_Saving_USD', 'sum')
).reset_index().sort_values(by='Total_Potential_Saving_USD', ascending=False)

print('=== Batching Opportunity by City ===')
print(city_summary.to_string(index=False))

# Bar chart of savings by city
plt.figure(figsize=(9, 5))
sns.barplot(data=city_summary, x='City', y='Total_Potential_Saving_USD', palette='viridis')
plt.title('Estimated Driver Cost Savings from Route Batching by City')
plt.ylabel('Potential Savings (USD)')
plt.xlabel('City')
for i, row in city_summary.reset_index().iterrows():
    plt.text(i, row['Total_Potential_Saving_USD'] + 200, f"${row['Total_Potential_Saving_USD']:,.0f}", ha='center', fontsize=9)
plt.tight_layout()
plt.show()
```

```python
# --- STEP 7: Interactive map of donor clusters ---
# Creates one map per city showing donors colored by batchable cluster
# Save as HTML files viewable in the browser

CITY_CENTERS = {
    'Boston':       [42.3601, -71.0589],
    'Chicago':      [41.8781, -87.6298],
    'New York City':[40.7128, -74.0060],
    'Los Angeles':  [34.0522, -118.2437]
}

for city, center in CITY_CENTERS.items():
    city_donors = geocoded[geocoded['Donor City'] == city]
    if city_donors.empty:
        continue

    m = folium.Map(location=center, zoom_start=14)

    # Plot all donors
    for _, row in city_donors.iterrows():
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=7,
            color='steelblue',
            fill=True,
            fill_color='steelblue',
            fill_opacity=0.7,
            popup=folium.Popup(row['Donor Name'], max_width=200)
        ).add_to(m)

    # Draw lines between batchable pairs in this city
    city_pairs = batch_df[batch_df['City'] == city]
    for _, pair in city_pairs.iterrows():
        a = city_donors[city_donors['Donor Name'] == pair['Donor_A']]
        b = city_donors[city_donors['Donor Name'] == pair['Donor_B']]
        if not a.empty and not b.empty:
            folium.PolyLine(
                locations=[
                    [a.iloc[0]['latitude'], a.iloc[0]['longitude']],
                    [b.iloc[0]['latitude'], b.iloc[0]['longitude']]
                ],
                color='tomato',
                weight=2,
                opacity=0.7,
                tooltip=f"{pair['Donor_A']} ↔ {pair['Donor_B']}: {pair['Distance_Miles']} mi"
            ).add_to(m)

    filename = f"{city.replace(' ', '_').lower()}_donor_map.html"
    m.save(filename)
    print(f'Saved: {filename}')

print('\nOpen any .html file in your browser to explore the interactive map.')
print('Blue dots = donor locations | Red lines = batchable pairs (≤0.5 miles apart)')
```
